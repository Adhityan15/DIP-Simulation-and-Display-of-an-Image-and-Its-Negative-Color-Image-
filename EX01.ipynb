```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read image
img = cv2.imread('input4.png')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Split RGB channels
R, G, B = cv2.split(img_rgb)

# Convert to grayscale
gray = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2GRAY)

# Negative of grayscale image
negative_gray = 255 - gray

# Convert grayscale to binary
_, binary = cv2.threshold(gray, 127, 1, cv2.THRESH_BINARY)

# Negative of binary image
negative_binary = 1 - binary

# Display images
plt.figure(figsize=(12, 10))

plt.subplot(3, 3, 1)
plt.title("Original Image")
plt.imshow(img_rgb)
plt.axis('off')

plt.subplot(3, 3, 2)
plt.title("Red Component")
plt.imshow(R, cmap='gray')
plt.axis('off')

plt.subplot(3, 3, 3)
plt.title("Green Component")
plt.imshow(G, cmap='gray')
plt.axis('off')

plt.subplot(3, 3, 4)
plt.title("Blue Component")
plt.imshow(B, cmap='gray')
plt.axis('off')

plt.subplot(3, 3, 5)
plt.title("Grayscale Image")
plt.imshow(gray, cmap='gray')
plt.axis('off')

plt.subplot(3, 3, 6)
plt.title("Negative Grayscale Image")
plt.imshow(negative_gray, cmap='gray')
plt.axis('off')

plt.subplot(3, 3, 7)
plt.title("Binary Image")
plt.imshow(binary, cmap='gray')
plt.axis('off')

plt.subplot(3, 3, 8)
plt.title("Negative Binary Image")
plt.imshow(negative_binary, cmap='gray')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read image
img = cv2.imread('input4.png')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Split channels
R, G, B = cv2.split(img_rgb)

# Create colored channel images
red_img = np.zeros_like(img_rgb)
red_img[:, :, 0] = R

green_img = np.zeros_like(img_rgb)
green_img[:, :, 1] = G

blue_img = np.zeros_like(img_rgb)
blue_img[:, :, 2] = B

# Negative of color image
negative_img = 255 - img_rgb

# Display
plt.figure(figsize=(12, 8))

plt.subplot(2, 3, 1)
plt.title("Original Color Image")
plt.imshow(img_rgb)
plt.axis('off')

plt.subplot(2, 3, 2)
plt.title("Red Component")
plt.imshow(red_img)
plt.axis('off')

plt.subplot(2, 3, 3)
plt.title("Green Component")
plt.imshow(green_img)
plt.axis('off')

plt.subplot(2, 3, 4)
plt.title("Blue Component")
plt.imshow(blue_img)
plt.axis('off')

plt.subplot(2, 3, 5)
plt.title("Negative Color Image")
plt.imshow(negative_img)
plt.axis('off')

plt.tight_layout()
plt.show()
```
